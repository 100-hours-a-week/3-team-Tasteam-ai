```python
import os, json
from openai import OpenAI
from transformers import pipeline

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
classifier = pipeline("text-classification", model="Dilwolf/Kakao_app-kr_sentiment")

data = json.load(open('/Users/js/tasteam-aicode-gpu-all-python-process-runtime_for_github/data/test_data_sample.json', 'r'))

target_idxs = []
reviews = []

# 1) restaurant_id==4ì¸ í•­ëª©ë§Œ ê³¨ë¼ì„œ (ì›ë³¸ dataì˜ indexë¥¼ ì €ì¥)
for i, review in enumerate(data):
    if review.get("restaurant_id") == 4:
        target_idxs.append(i)
        reviews.append(review["content"])

# 2) LLMì— "index í¬í•¨"í•´ì„œ ë³´ë‚´ê¸° (indexëŠ” reviews ë‚´ë¶€ index)
llm_input = "\n".join([f"{j}\t{txt}" for j, txt in enumerate(reviews)])

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "You are a sentiment analysis expert.\n"
                "Classify each numbered Korean review as positive, negative, or neutral.\n"
                "Return ONLY a valid JSON array.\n"
                'Format: [{"index": 0, "sentiment": "positive|negative|neutral"}]'
            )
        },
        {"role": "user", "content": llm_input}
    ],
    temperature=0
)

results = json.loads(response.choices[0].message.content)

# 3) ê²°ê³¼ë¥¼ ì›ë³¸ dataì— sentimentë¡œ ì£¼ì…
for r in results:
    j = r["index"]                 # reviews ë‚´ë¶€ index
    sentiment = r["sentiment"]
    original_i = target_idxs[j]    # ì›ë³¸ data index
    data[original_i]["sentiment"] = sentiment

# í™•ì¸
print(data[target_idxs[0]]["content"])
print(data[target_idxs[0]]["sentiment"])
```
