```python
def parse_reviews_from_file(path):
    reviews = []
    buffer = []

    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()

            if not line:  # 공백 줄 → 리뷰 종료
                if buffer:
                    reviews.append(" ".join(buffer))
                    buffer = []
            else:
                buffer.append(line)

    if buffer:
        reviews.append(" ".join(buffer))

    return reviews


path = "/Users/js/tasteam-aicode-gpu-all-python-process-runtime_for_github/hybrid_search/data_preprocessing/yul_review.txt"
reviews = parse_reviews_from_file(path)

print(len(reviews))
print(reviews[:3])
```
