```python
from fastembed import TextEmbedding
import pandas as pd
import numpy as np

supported_text_models = (
    pd.DataFrame(TextEmbedding.list_supported_models())
    .sort_values("size_in_GB")
    .drop(columns=["sources","model_file","additional_files"])
    .reset_index(drop=True)
)
supported_text_models
```
